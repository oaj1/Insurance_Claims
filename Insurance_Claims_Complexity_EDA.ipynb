```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#lets take a look at the data, fraud vs non fraud as it relates to number of vehicles involved
df = pd.read_csv('insurance_claims_cleaned_fixed.csv')


#lets take a summary look at the data for number of vhicles involved, grouped by fraud
summary = df.groupby('fraud_reported')['number_of_vehicles_involved'].describe()

print(summary)
```

```python
plt.figure(figsize=(10,6))
custom_palette = {'No': 'orange', 'Yes': 'blue'}
ax = sns.boxplot(
    x='fraud_reported',
    y='number_of_vehicles_involved',
    data=df,
    hue='fraud_reported',
    palette=custom_palette,
    #legend=False
)
plt.title('Number of Vehicles Involved by Fraud Reported')
plt.xlabel('Fraud Reported')
plt.ylabel('Number of Vehicles Involved')
# Add legend manually
handles = [plt.Line2D([0], [0], color='blue', lw=4), plt.Line2D([0], [0], color='orange', lw=4)]
labels = ['Yes', 'No']
plt.legend(handles, labels, title='Fraud Reported', loc='upper right')
plt.show()
```


In summary: Fraudulent claims tend to involve a slightly higher average number of vehicles, but the overall distribution is very similar between fraud and non-fraud cases. Most claims, regardless of fraud status, involve just one vehicle. The number of vehicles involved does not appear to be a strong standalone predictor of fraud. 

```python
#now lets take a look at bodily injury vs fraud, grouped by fraud
bodily_injury_summary = df.groupby('fraud_reported')['bodily_injuries'].describe()
print(bodily_injury_summary)
```

```python
plt.figure(figsize=(10,6))
custom_palette = {'No': 'orange', 'Yes': 'blue'}
ax = sns.boxplot(
    x='fraud_reported',
    y='bodily_injuries',
    data=df,
    hue='fraud_reported',
    palette=custom_palette,
    #legend=False
)
plt.title('Number of Bodily Injuries by Fraud Reported')
plt.xlabel('Fraud Reported')
plt.ylabel('Number of Bodily Injuries')
# Add legend manually
handles = [plt.Line2D([0], [0], color='blue', lw=4), plt.Line2D([0], [0], color='orange', lw=4)]
labels = ['Yes', 'No']
plt.legend(handles, labels, title='Fraud Reported', loc='upper right')
plt.show()
```

### Summary of Bodily Injuries by Fraud Reported
- **Non-fraudulent claims** (No):
  - Count: 753
  - Mean bodily injuries: 0.98
  - Most claims have 1 or fewer injuries (median = 1.0)
  - 75% of claims have 2 or fewer injuries
- **Fraudulent claims** (Yes):
  - Count: 247
  - Mean bodily injuries: 1.04
  - Most claims have 1 or fewer injuries (median = 1.0)
  - 75% of claims have 2 or fewer injuries

**Conclusion:**
The average number of bodily injuries is slightly higher in fraudulent claims, but the overall distribution is very similar between fraud and non-fraud cases.

**Insight:**
The number of bodily injuries does not appear to be a strong standalone predictor of fraud.
